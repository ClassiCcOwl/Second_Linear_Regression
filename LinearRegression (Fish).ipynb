```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("Fish.csv")
data
```

```python
data.drop("Species", axis=1, inplace=True)
data
```

```python
df = pd.DataFrame(data)
df
```

```python
W_max = df["Weight"].nlargest(5)
df.drop(W_max.index, axis = 0,inplace=True)
df.reset_index(inplace = True, drop=True)
```

```python
l_min = df["Length2"].nsmallest(2)
df.drop(l_min.index, axis = 0,inplace=True)
df.reset_index(inplace = True, drop=True)
```

```python
df_Weight= pd.DataFrame(df, columns = ["Weight"])
df_Weight.sort_values(by=["Weight"],inplace=True)
df_Weight.reset_index(inplace = True, drop=True)
df_Weight
plt.plot(df_Weight["Weight"])
plt.show()
```

```python
under_30 = df[df["Weight"]<20]
df.drop(under_30.index, axis = 0,inplace=True)
df.reset_index(inplace = True, drop=True)
```

```python
df_Weight= pd.DataFrame(df, columns = ["Weight"])
df_Weight.sort_values(by=["Weight"],inplace=True)
df_Weight.reset_index(inplace = True, drop=True)
df_Weight
plt.plot(df_Weight["Weight"])
plt.show()
```

```python
df.describe()
```

```python
X = pd.DataFrame(df , columns = ["Length1", "Length2", "Length3", "Height", "Width"])
y = df["Weight"].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y , test_size = 0.2 , random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
print(regressor.intercept_ , regressor.coef_)
```

```python
print("R2 Score:", metrics.r2_score(y_test, y_pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Mean Absoulute Error:", metrics.mean_absolute_error(y_test, y_pred))
```

```python
fig ,(ax1,ax2,ax3) = plt.subplots(3 ,figsize=(5,8))
fig.tight_layout()
ax1.set_title("Length1 / Weight")
ax1.scatter(X_test.Length1, y_test)
ax1.scatter(X_test.Length1, y_pred)
ax1.grid()

ax2.set_title("Length2 / Weight")
ax2.scatter(X_test.Length2, y_test)
ax2.scatter(X_test.Length2, y_pred)
ax2.grid()

ax3.set_title("Length3 / Weight")
ax3.scatter(X_test.Length3, y_test)
ax3.scatter(X_test.Length3, y_pred)
ax3.grid()
```

```python
fig ,(ax1,ax2) = plt.subplots(2 ,figsize=(5,8))
fig.tight_layout()
ax1.set_title("Height / Weight")
ax1.scatter(X_test.Height, y_test)
ax1.scatter(X_test.Height, y_pred)
ax1.grid()

ax2.set_title("Width / Weight")
ax2.scatter(X_test.Width, y_test)
ax2.scatter(X_test.Width, y_pred)
ax2.grid()
```
